```python
# function to add extension to "id column"
def append_ext(fn):
    return fn+".jpg"
```

```python
# read training dataframe provided with labels
train_df = pd.read_csv("labels/labels.csv")
train_df.head()
```

```python
# read submission test dataframe
test_df = pd.read_csv("sample submission/sample_submission.csv")
test_df.head()
```

```python
# apply extensions to train_df and test_df
train_df["id"]=train_df["id"].apply(append_ext)
test_df["id"]=test_df["id"].apply(append_ext)
```

```python
# validate "id" column .jpg extension was added
train_df.head()
```

```python
# Groupby breeds and count the number of images per breed 
# we have in our training set
train_df_sorted = train_df.groupby(train_df["breed"]).count()
```

```python
# sort by highest represented breeds
train_df_sorted = train_df_sorted.sort_values(by=["id"], ascending=False)
train_df_sorted.head(8)
```
